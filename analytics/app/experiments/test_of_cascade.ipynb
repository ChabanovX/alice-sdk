```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

from ml_models import BiClassifier, MultiClassifier, BertEmbedder

BIN_THRESH = 0.80
RANDOM_STATE = 42
N_FOLDS = 5
DATA_1_PATH = "/Users/rami/Documents/driver_intention_classifier/datasets/dataset_6k_10_classes.csv"
DATA_2_PATH = "/Users/rami/Documents/driver_intention_classifier/datasets/dataset_other_12k.csv"

print("Loading data and creating embeddings…")
data_1 = pd.read_csv(DATA_1_PATH, sep=";")
data_2 = pd.read_csv(DATA_2_PATH, sep=",").dropna().sample(frac=0.65, random_state=RANDOM_STATE).reset_index(drop=True)
data = pd.concat([data_1, data_2]).reset_index(drop=True)
data["label"] = data["label"].fillna("other").astype(str)
y_bin = (data["label"] != "other").astype(int)
y_mult = data["label"]

embedder = BertEmbedder()
X = np.vstack([embedder.embed(t) for t in data["text"]])
print(f"Embeddings ready: X.shape = {X.shape}\n")

skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=RANDOM_STATE)
y_pred_bin_cv = np.zeros(len(X), dtype=int)
y_pred_mult_cv = np.array(["other"] * len(X), dtype=object)
fold_acc = []

for fold, (tr_idx, te_idx) in enumerate(skf.split(X, y_bin), 1):
    bin_clf = BiClassifier(bin_thresh=BIN_THRESH, iterations=500, learning_rate=0.05, random_seed=RANDOM_STATE, verbose=False).fit(X[tr_idx], y_bin.iloc[tr_idx])
    y_hat_bin = bin_clf.predict(X[te_idx])
    y_pred_bin_cv[te_idx] = y_hat_bin
    print(f"\n—— Binary report (fold {fold}) ——")
    print(classification_report(y_bin.iloc[te_idx], y_hat_bin, target_names=["other(0)", "inlier(1)"]))
    sns.heatmap(confusion_matrix(y_bin.iloc[te_idx], y_hat_bin), annot=True, fmt="d", cmap="Blues", xticklabels=["other", "inlier"], yticklabels=["other", "inlier"])
    plt.title(f"Confusion matrix — binary fold {fold}")
    plt.show()
    mask_tr = y_bin.iloc[tr_idx].values == 1
    mask_te = y_hat_bin == 1
    X_tr_inl, y_tr_inl = X[tr_idx][mask_tr], y_mult.iloc[tr_idx][mask_tr]
    X_te_inl, y_te_inl = X[te_idx][mask_te], y_mult.iloc[te_idx][mask_te]
    keep_tr = y_tr_inl != "other"
    keep_te = y_te_inl != "other"
    X_tr_inl, y_tr_inl = X_tr_inl[keep_tr], y_tr_inl[keep_tr]
    X_te_inl, y_te_inl = X_te_inl[keep_te], y_te_inl[keep_te]
    if len(y_tr_inl) > 0:
        multi_clf = MultiClassifier(n_bags=4, depths=[6, 8], lrs=[0.03, 0.05], regs=[3, 5], iterations=500, random_state=fold * 10, verbose=False).fit(X_tr_inl, y_tr_inl)
        y_hat_mult = multi_clf.predict_many(X_te_inl)
        y_pred_mult_cv[te_idx[mask_te][keep_te]] = y_hat_mult
        lbls = sorted(pd.unique(y_tr_inl))
        print(f"\n—— Multiclass report (fold {fold}) ——")
        print(classification_report(y_te_inl, y_hat_mult, target_names=lbls))
        sns.heatmap(confusion_matrix(y_te_inl, y_hat_mult, labels=lbls), annot=True, fmt="d", cmap="Blues", xticklabels=lbls, yticklabels=lbls)
        plt.title(f"Confusion matrix — multi fold {fold}")
        plt.show()
    y_pred_fold = y_pred_mult_cv[te_idx].copy()
    y_pred_fold[y_hat_bin == 0] = "other"
    fold_acc.append(accuracy_score(y_mult.iloc[te_idx], y_pred_fold))
    print(f"Cascade accuracy (fold {fold}) = {fold_acc[-1]:.4f}")

y_pred_cascade = y_pred_mult_cv.copy()
y_pred_cascade[y_pred_bin_cv == 0] = "other"
cascade_acc = accuracy_score(y_mult, y_pred_cascade)
print(f"\nOverall cascade CV accuracy = {cascade_acc:.4f} (mean ± std per fold: {np.mean(fold_acc):.4f} ± {np.std(fold_acc):.4f})")

print("\nFitting final binary model on ALL data…")
final_bin = BiClassifier(bin_thresh=BIN_THRESH, iterations=500, learning_rate=0.05, random_seed=RANDOM_STATE, verbose=False).fit(X, y_bin)
print("Fitting final multiclass model on ALL inlier data…")
inlier_mask = y_bin.values == 1
final_multi = MultiClassifier(n_bags=4, depths=[6, 8], lrs=[0.03, 0.05], regs=[3, 5], iterations=500, random_state=RANDOM_STATE, verbose=False).fit(X[inlier_mask], y_mult[inlier_mask])
print("Both final models are ready.\n")

def predict_binary(text):
    vec = embedder.embed(text).reshape(1, -1)
    return "inlier" if final_bin.predict(vec)[0] == 1 else "other"

def predict_multi(text):
    vec = embedder.embed(text).reshape(1, -1)
    if final_bin.predict(vec)[0] == 0:
        return "other"
    return final_multi.predict_many(vec)[0]
```
